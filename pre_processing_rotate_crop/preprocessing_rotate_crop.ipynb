```python
import skimage
import matplotlib.pyplot as plt
from skimage.io import imread, imshow, imsave
from skimage.transform import rotate
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import glob
%matplotlib inline  
```

```python
bs=300
imagefile=glob.glob('test/*.JPG')

directory='test_crop'
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
imagefile
```

```python
for file in imagefile:
    try:
        print(file)
        # read the image
        img=cv2.imread(file)
        
        file = file.replace('\\', '/')
        
        imagename=file.split('/')[-1]
        img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
        # approximated mean values of h s v
        hsv=[6,115,200]
        # approximated bounds of h s v
        thresh=[20,20,20]
        # lower bound
        minHSV = np.array([hsv[0] - thresh[0], hsv[1] - thresh[1], hsv[2] - thresh[2]])
        # uppper bound
        maxHSV = np.array([hsv[0] + thresh[0], hsv[1] + thresh[1], hsv[2] + thresh[2]])
        img1 = cv2.inRange(img_hsv, minHSV, maxHSV)



        # rotate image
        pixels = np.argwhere(img1 == 255)
        nx, ny = np.shape(img1)
        meanx=np.mean(pixels[:,0])
        meany=np.mean(pixels[:,1])


        if nx<ny:
            if meany<ny/2:
                # 90 degrees
                rotated = cv2.rotate(img, cv2.ROTATE_90_CLOCKWISE)
                rotated2 = rotated[int(meany+300):-bs,bs:-bs,:]
            elif meany>ny/2:
                # 90 degrees
                rotated = cv2.rotate(img, cv2.ROTATE_90_COUNTERCLOCKWISE)
                rotated2 = rotated[int(ny-meany+300):-bs,bs:-bs,:]
        if ny<nx:
            if meanx>nx/2:
                # 180 degrees
                rotated = cv2.rotate(img, cv2.ROTATE_180)
                rotated2 = rotated[bs:int(meanx-300-bs),bs:-bs,:]
            elif meanx<nx/2:
                rotated2 = img[int(meanx+bs):-bs,bs:-bs,:]
        img_rgb = cv2.cvtColor(rotated2, cv2.COLOR_BGR2RGB)
        # save image  
        cv2.imwrite(directory + '/'  + imagename, rotated2)
    except:
        print("An exception occurred")
```
